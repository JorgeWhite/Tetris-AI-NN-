```python
### Implementamos la clase del tablero del juego que se encargara de la logica.

class Board():
    def __init__(self):
        ## Creating the grid
        ## X - 10 blocks
        ## Y - 20 blocks
        self.grid = np.array([[0]*10]*20)
        self.auxGrid = np.array([[0]*10]*20)
        self.spawnZone = self.auxGrid[0:4]
        self.pieceZone = self.grid[0:4]
        self.pieceZoneX = 3
        self.pieceZoneY = 0
        self.points = 0
        self.introducePiece(self.getRandomPiece())
        #self.introducePiece(OPiece())
        self.nextPiece = self.getRandomPiece()
        #self.nextPiece = OPiece()
        self.isGameOver = False

    def updateZones(self):
        self.pieceZoneX = 3
        self.pieceZoneY = 0

    def showBoard(self):
        print(self.auxGrid)

    def introducePiece(self, piece):
        self.activePiece = piece
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[i][j] == 1:
                    if self.grid[i][3 + j] != 1:
                        aux += 1
        if aux == 4:
            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[i][j] == 1:
                        self.auxGrid[i][3 + j] = 1
        else:
            #print("Game Over")
            self.isGameOver = True

    def checkGameOver(self):
        for i in range(4):
            for j in range(4):
                if self.spawnZone[i][j + 3] == 1 and self.grid[self.pieceZoneY + 1][self.pieceZoneX + j] == 1:
                    #print("Game Over")
                    self.isGameOver = True

    def updateBoard(self, piece):
        self.pieceZone[0][3:7] = piece.shape[0]
        self.pieceZone[1][3:7] = piece.shape[1]
        self.pieceZone[2][3:7] = piece.shape[2]
        self.pieceZone[3][3:7] = piece.shape[3]

    def movePieceRight(self):
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][3 - i] == 1:
                    if self.pieceZoneX + 3 - i + 1 >= 10:
                        #print("There's a wall")
                        return
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][3 - i] == 1:
                    if self.grid[self.pieceZoneY + j][self.pieceZoneX + 3 - i + 1] != 1:
                        aux += 1
        if aux == 4:

            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[j][3 - i] == 1:
                        self.auxGrid[self.pieceZoneY + j][self.pieceZoneX + 3 - i + 1] = 1
                        self.auxGrid[self.pieceZoneY + j][self.pieceZoneX + 3 - i] = 0

            self.pieceZoneX += 1
        #else:
            #print("There's another piece")

    def movePieceLeft(self):
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][i] == 1:
                    if self.pieceZoneX + i - 1 <= -1:
                        #print("there's a wall")
                        return
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][i] == 1:
                    if self.grid[self.pieceZoneY + j][self.pieceZoneX + i - 1] != 1:
                        aux += 1
        if aux == 4:

            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[j][i] == 1:
                        self.auxGrid[self.pieceZoneY + j][self.pieceZoneX + i - 1] = 1
                        self.auxGrid[self.pieceZoneY + j][self.pieceZoneX + i] = 0
            self.pieceZoneX -= 1
        #else:
            #print("There's another piece")

    def moveSoftDrop(self):
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[3 - i][j] == 1:
                    if self.pieceZoneY + 3 - i + 1 >= 20:
                        #print("the piece is at the bottom. it'll be stacked")
                        self.stackPiece()
                        return
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[3 - i][j] == 1:
                    if self.grid[self.pieceZoneY + 3 - i + 1][self.pieceZoneX + j] != 1:
                        aux += 1
        if aux == 4:
            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[3 - i][j] == 1:
                        self.auxGrid[self.pieceZoneY + 3 - i + 1][self.pieceZoneX + j] = 1
                        self.auxGrid[self.pieceZoneY + 3 - i][self.pieceZoneX + j] = 0
            self.pieceZoneY += 1
            self.points += 1
        else:
            #print("There´s another piece, it'll be stacked")
            self.stackPiece()
            
    def movePieceDown(self):
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[3 - i][j] == 1:
                    if self.pieceZoneY + 3 - i + 1 >= 20:
                        #print("the piece is at the bottom. it'll be stacked")
                        self.stackPiece()
                        return
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[3 - i][j] == 1:
                    if self.grid[self.pieceZoneY + 3 - i + 1][self.pieceZoneX + j] != 1:
                        aux += 1
        if aux == 4:
            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[3 - i][j] == 1:
                        self.auxGrid[self.pieceZoneY + 3 - i + 1][self.pieceZoneX + j] = 1
                        self.auxGrid[self.pieceZoneY + 3 - i][self.pieceZoneX + j] = 0
            self.pieceZoneY += 1
            self.points += 0
        else:
            #print("There´s another piece, it'll be stacked")
            self.stackPiece()

    def rotatePieceCW(self):
        self.activePiece.rotateRight()
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][3 - i] == 1:
                    if self.pieceZoneX + 3 - i + 1 >= 11 or self.pieceZoneX + i - 1 <= -2:
                        #print("No wall kick")
                        self.activePiece.rotateLeft()
                        return
        aux = 0
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[i][j] == 1 and self.pieceZoneX + j < 10 and self.pieceZoneY + i < 20:
                    if self.grid[self.pieceZoneY + i][self.pieceZoneX + j] != 1:
                        aux += 1
        if aux == 4:
            self.auxGrid = np.array(self.grid)
            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[i][j] == 1:
                        self.auxGrid[self.pieceZoneY + i][self.pieceZoneX + j] += self.activePiece.shape[i][j]
        else:
            self.activePiece.rotateLeft()
            #print("It can't rotate")

    def rotatePieceCCW(self):
        aux = 0
        self.activePiece.rotateLeft()
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[j][3 - i] == 1:
                    if self.pieceZoneX + 3 - i + 1 >= 11 or self.pieceZoneX + i - 1 <= -2:
                        #print("No wall kick")
                        self.activePiece.rotateRight()
                        return
        for i in range(4):
            for j in range(4):
                if self.activePiece.shape[i][j] == 1 and self.pieceZoneX + j < 10 and self.pieceZoneY + i < 20:
                    if self.grid[self.pieceZoneY + i][self.pieceZoneX + j] != 1:
                        aux += 1
        if aux == 4:
            self.auxGrid = np.array(self.grid)
            for i in range(4):
                for j in range(4):
                    if self.activePiece.shape[i][j] == 1:
                        self.auxGrid[self.pieceZoneY + i][self.pieceZoneX + j] += self.activePiece.shape[i][j]
        else:
            self.activePiece.rotateRight()
            #print("It can't rotate")

    def stackPiece(self):
        self.grid = np.array(self.auxGrid)
        self.updateZones()
        self.checkCompletedLines()
        self.introducePiece(self.nextPiece)
        self.nextPiece = self.getRandomPiece()
        #self.nextPiece = OPiece()

    def checkCompletedLines(self):
        linesCompleted = 0
        for i in range(20):
            if sum(self.grid[i]) == 10:
                self.updateLines(i)
                linesCompleted += 1
        if linesCompleted != 0:
            print(linesCompleted, " lines were completed")
        if linesCompleted == 1:
            self.points += 760
        elif linesCompleted == 2:
            self.points += 1900
        elif linesCompleted == 3:
            self.points += 5700
        elif linesCompleted == 4:
            self.points += 22800


    def moveMinoRight(self, y, x):
        if x <= 8:
            if self.auxGrid[y][x] == 1:
                self.auxGrid[y][x + 1] = 1
                self.auxGrid[y][x] = 0

    def updateLines(self, row):
        for i in range(row - 1):
            self.auxGrid[row - i] = self.auxGrid[row - i - 1]
        self.grid = np.array(self.auxGrid)

    def getRandomPiece(self):
        random = np.random.rand()
        if 0 < random < 1 / 7:
            l = LPiece()
            return l
        elif 1 / 7 < random < 2 / 7:
            j = JPiece()
            return j
        elif 2 / 7 < random < 3 / 7:
            s = SPiece()
            return s
        elif 3 / 7 < random < 4 / 7:
            z = ZPiece()
            return z
        elif 4 / 7 < random < 5 / 7:
            o = OPiece()
            return o
        elif 5 / 7 < random < 6 / 7:
            i = IPiece()
            return i
        else:
            t = TPiece()
            return t

    def getNextPiecePreview(self):
        nextPiece = self.getRandomPiece()
        return nextPiece.shape
    
    def quitGame(self):
        sys.exit()

    def showBoard2(self):
        self.boardShow = np.array([[str]*10]*20)
        self.nextPiecePreview = np.array([[str]*4]*4)
        for i in range(20):
            for j in range(10):
                if self.auxGrid[i][j] == 1:
                    self.boardShow[i][j] = "o︎"
                else:
                    self.boardShow[i][j] = " "
        for i in range(4):
            for j in range(4):
                if self.nextPiece.shape[i][j] == 1:
                    self.nextPiecePreview[i][j] = "o"
                else:
                    self.nextPiecePreview[i][j] = ' '

        print(self.boardShow)
        print("Next Piece:")
        print(self.nextPiecePreview), print("Points: ", self.points)
```

```python
### Creamos la clase del juego.
class game():
    def __init__(self):
        self.board = Board()
        self.running = True
    def action(self):
            move = input()
            if (move == 'q'):
                self.running = False
                return
            if move == 'a':
                self.board.movePieceLeft()
                self.board.showBoard2()
                return
            if move == 'd':
                self.board.movePieceRight()
                self.board.showBoard2()
                return
            if move == 's':
                self.board.moveSoftDrop()
                self.board.showBoard2()
                return
            if move == 'g':
                self.board.rotatePieceCCW()
                self.board.showBoard2()
                return
            if move == 'h':
                self.board.rotatePieceCW()
                self.board.showBoard2()
                return
            return

    def run(self):
        while self.running == True and self.board.isGameOver == False:
            self.action()
            
    def state(self):
        array = []
        for i in range(20):
            for j in range(10):
                array.append(self.board.auxGrid[i][j])
        return array
    
    def makeMove(self, OHE_array):
        if OHE_array[0] == 1:
            #print("NN left")
            self.board.movePieceLeft()
        elif OHE_array[1] == 1:
            #print("NN rigth")
            self.board.movePieceRight()
        elif OHE_array[2] == 1:
            #print("NN down")
            self.board.moveSoftDrop()
        elif OHE_array[3] == 1:
            #print("NN CCW")
            self.board.rotatePieceCCW()
        elif OHE_array[4] == 1:
            #print("NN CW")
            self.board.rotatePieceCW()
        elif OHE_array[5] == 1:
            #print("NN nothing")
            return

    def receiveFromNN(self, NN = None):
        move = NN.forward(self.state())
        move = self.makeY(move)
        self.makeMove(move)

    
    def pointsFromNN(self):
        return self.board.points
    
    ### Procesamos la salida de la red para que pueda dar un input al juego.
    def makeY(self, OHE_tensor):
        out_OHE = np.array([0]*6)
        ind = torch.argmax(OHE_tensor)
        out_OHE[ind] = 1
        return out_OHE


```
